```python
def InverseInnerDict( dictionary ):
    new_dict = {}
    for key in dictionary:
        new_inner_dict = {}
        inner_dict = dictionary[ key ]
        for inner_key in inner_dict:
            if ( type( inner_dict[ inner_key ] ) == list ):
                inner_dict[ inner_key ] = tuple( inner_dict[ inner_key ] )
            new_inner_dict[ inner_dict[ inner_key ] ] = inner_key
        new_dict[ key ] = new_inner_dict
    return new_dict
```

```python
def pickleThing( thing, to_filename ):
    file = open( to_filename, "ab" )
    pickle.dump( thing, file )
    file.close()
```
